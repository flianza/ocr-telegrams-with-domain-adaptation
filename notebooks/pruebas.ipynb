```python
import os
from collections import defaultdict
from matplotlib import pyplot as plt
from PIL import Image
import cv2 
```

```python
root_path = '/Users/francolianza/code/medgc-tesis/data/05_model_input/TDS/tds_val_image'
files = [f"{root_path}/{file}" for file in os.listdir(root_path) if file.endswith(".png")]
```

```python
fig, axs = plt.subplots(10, 10, figsize=(20,20))

for i, (file, ax) in enumerate(zip(files, axs.flatten())):
    img = cv2.imread(file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.bitwise_not(img)
    ax.imshow(img)
    ax.axis('off')
```

```python
path = '/Users/francolianza/code/medgc-tesis/data/05_model_input/TDS/image_list/tds_{}.txt'
d = {'train': [], 'test': [], 'val': []}
for phase in d.keys():
    with open(path.format(phase), "r") as f:
        d[phase] = f.readlines()
```
